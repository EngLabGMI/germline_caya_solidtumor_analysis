```r
df_to_MutationMapper =  function( gene_name  , df ){
    df %>% dplyr::filter( Hugo_Symbol == gene_name ) %>% dplyr::rename( "Mutation_Type" = Variant_Classification)
}
```

```r
palb2 = df_to_MutationMapper('PALB2', df)
```

```r
write.table(palb2, "../ALL_Variants/maf_res/palb2_data_for_lolipop.txt", row.names = F, quote = F, sep = "\t"  )
```

```r
gene = df_to_MutationMapper('TP53', df)
write.table(gene, "../ALL_Variants/maf_res/TP53_data_for_lolipop.txt", row.names = F, quote = F, sep = "\t"  )
```
