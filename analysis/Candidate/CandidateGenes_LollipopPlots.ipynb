```r
df_to_MutationMapper =  function( gene_name  , df ){
    df %>% dplyr::filter( Hugo_Symbol == gene_name ) %>% dplyr::rename( "Mutation_Type" = Variant_Classification)
}
```

```r
gene = df_to_MutationMapper('CRIPAK', df)
write.table(gene, "../ALL_Variants/maf_res/CRIPAK_data_for_lolipop.txt", row.names = F, quote = F, sep = "\t"  )
```

```r
gene = df_to_MutationMapper('PRKN', df)
write.table(gene, "../ALL_Variants/maf_res/PRKN_data_for_lolipop.txt", row.names = F, quote = F, sep = "\t"  )
```
